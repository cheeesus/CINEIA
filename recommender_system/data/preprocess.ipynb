```python
import pandas as pd
from load_data import movies_df, genres_df, keywords_df, movies_genre_df, movies_keyword_df
```

```python
# Afficher les premières lignes des DataFrames
print("Movies:")
print(movies_df.head())

print("\nGenres:")
print(genres_df.head())

print("\nKeywords:")
print(keywords_df.head())

print("\nMovie Genres:")
print(movies_genre_df.head())

print("\nMovie Keywords:")
print(movies_keyword_df.head())

# Obtenir des informations sur les DataFrames
print("\nMovies info:")
print(movies_df.info())

print("\nGenres info:")
print(genres_df.info())

# Statistiques descriptives
print("\nMovies statistics:")
print(movies_df.describe())

# Valeurs uniques dans les colonnes
print("\nUnique genres:")
print(genres_df['name'].unique())

print("\nUnique keywords:")
print(keywords_df['name'].unique())

#Nombre de films par genres
movies_genre_counts = movies_genre_df['genre_id'].value_counts()
print("\nMovies per genre:")
print(movies_genre_counts)
```

```python
# Fusionner movies_genre avec genre pour récupérer le nom du genre
movies_genre_df = movies_genre_df.merge(genres_df, left_on="genre_id", right_on="id", how="left")
movies_genre_df = movies_genre_df.groupby("movie_id")["name"].apply(lambda x: " ".join(x)).reset_index()

# Fusionner movies_keyword avec keywords pour récupérer le nom du mot-clé
movies_keyword_df = movies_keyword_df.merge(keywords_df, left_on="keyword_id", right_on="id", how="left")
movies_keyword_df = movies_keyword_df.groupby("movie_id")["name"].apply(lambda x: " ".join(x)).reset_index()
```

```python
# Sauvegarder movies_df en CSV
movies_df.to_csv('database_data/movies.csv', index=False)

# Sauvegarder genres_df en CSV
genres_df.to_csv('database_data/genres.csv', index=False)

# Sauvegarder keywords_df en CSV
keywords_df.to_csv('database_data/keywords.csv', index=False)

# Sauvegarder movies_genre_df en CSV
movies_genre_df.to_csv('database_data/movies_genre.csv', index=False)

# Sauvegarder movies_keyword_df en CSV
movies_keyword_df.to_csv('database_data/movies_keyword.csv', index=False)

print("Tous les DataFrames ont été sauvegardés avec succès dans le dossier database_data/")
```
